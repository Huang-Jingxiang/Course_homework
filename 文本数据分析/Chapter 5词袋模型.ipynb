# 第一次作业 词袋模型
## 英文文本

```python
import pandas as pd
text_ana = pd.read_excel('text_analysis_twitter_sample.xlsx')
text_ana = text_ana.drop(['Unnamed: 0'], axis=1)
text_ana
```

```python
import re
def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'http\S+', '', text)
    # 去除中文标点符号
    text = re.sub(r'，', '', text)
    stop_words = set(['to', 'the', 'for', 'and', 'of', 'this', 'in', 'on', 'rt', 'is', 'amp', 'our', 'you', 'we', 'with', 'at'
                      ,'it', 'qt', 'my', 'be', 'that', 'are', 'from', 'as', 'have', 'by', '&amp'])
    text = ' '.join([word for word in text.split() if word not in stop_words])
    return text

text_ana['text'] = text_ana['text'].apply(preprocess_text)
# text_ana['text'].apply(preprocess_text)
```

```python
text_ana
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()

X = vectorizer.fit_transform(text_ana['text'])

print('shape:', X.shape)

print('feature list:', vectorizer.get_feature_names_out()[900:1000])
```

```python
for i in range(X.shape[0]):
    print(X[i].toarray())
```

```python
feature_words = vectorizer.get_feature_names_out()

word_freq = dict(zip(feature_words, X.sum(axis=0).A1))
```

```python
sorted(word_freq.items(), key=lambda x: x[1], reverse=True)
```

## 中文文本

```python
text_ana_zh = pd.read_excel('text_analysis_weibo_sample.xlsx')
text_ana_zh = text_ana_zh.drop(['Unnamed: 0'], axis=1)
text_ana_zh
```

```python
import jieba
text_ana_zh['标题/微博内容'] = text_ana_zh['标题/微博内容'].apply(preprocess_text)

def cut_words(text):
    return ' '.join(jieba.cut(text))
text_ana_zh['分词'] = text_ana_zh['标题/微博内容'].apply(cut_words)
```

```python
text_ana_zh
```

```python
vectorizer = CountVectorizer()

X = vectorizer.fit_transform(text_ana_zh['分词'])

print('shape:', X.shape)

print('feature list:', vectorizer.get_feature_names_out()[900:1000])
```

```python
for i in range(X.shape[0]):
    print(X[i].toarray())
```

```python
feature_words = vectorizer.get_feature_names_out()

word_freq = dict(zip(feature_words, X.sum(axis=0).A1))

sorted (word_freq.items(), key=lambda x: x[1], reverse=True)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import matplotlib


default_font_path = matplotlib.font_manager.FontProperties(family='DejaVu Sans').get_file()


wc = WordCloud(
    width=1200,
    height=800,
    max_words=200,
    colormap='Blues',
    background_color='white',
    font_path='SimSun.ttf'
).generate_from_frequencies(word_freq)


# plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']
plt.figure(figsize=(10,5))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
